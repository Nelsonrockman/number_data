```
import cv2

def read_process(images,x,y,dir_name):
  for j in images:
    img=cv2.imread(str(dir_name)+"/"+j)
    img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img=cv2.resize(img, (32, 32)) 
    x.append(img)
    y.append(dir_name)

import os

x=[]#appending images
y=[]#appending labels
for i in range(0,10):
  l=os.listdir(str(i))
  read_process(l,x,y,i)
print("total images",len(x))
print("total labels",len(y))
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```
import numpy as np
x_train=np.array(x_train)
x_test=np.array(x_test)
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```
x_train = x_train.reshape((x_train.shape[0], 32, 32, 1))
x_test = x_test.reshape((x_test.shape[0], 32, 32, 1))
```

```
train_norm = x_train.astype('float32')
test_norm = x_test.astype('float32')
# normalize to range 0-1
train_norm = train_norm / 255.0
test_norm = test_norm / 255.0
import tensorflow as tf
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)
```

```
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Dense
from keras.layers import Flatten
from keras.optimizers import SGD
from keras.layers import Dropout
  
  
## Declare the model
model = Sequential()

## Declare the layers
layer_1 = Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(32, 32, 1))
layer_2 = Conv2D(64, kernel_size=(3,3), activation='relu')
layer_3 = Conv2D(128, kernel_size=(3,3), activation='relu')
layer_4 = MaxPooling2D(pool_size=(2,2))
layer_5 = MaxPooling2D(pool_size=(2,2))
layer_6 = Dropout(0.75)
layer_7 = Flatten()
#layer_6 = Dense(30, activation='relu')
layer_8 = Dropout(0.25)
layer_9 = Dense(10, activation='softmax')

## Add the layers to the model
model.add(layer_1)
model.add(layer_2)
model.add(layer_3)
model.add(layer_4)
model.add(layer_5)
model.add(layer_6)
model.add(layer_7)
model.add(layer_8)
model.add(layer_9)
```

```
model.compile(optimizer="adam", loss='categorical_crossentropy', metrics=['accuracy'])
```

```
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=25)
```
